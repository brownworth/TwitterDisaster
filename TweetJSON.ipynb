Pandas is the python library for panel data. It allows you to create dataframes and series out of tabular data, like json. bringing it in as 'pd' is just a common shortcut.

re is the python library for regular expressions

```python
import pandas as pd
import re
```

Here we're reading the json into a pandas dataframe. Since it is on multiple lines, we add the `lines=True` part to avoid an error

```python
tweet_json = pd.read_json(r'./top10-chiacago.txt',lines=True)
```

here's the data in that new variable

```python
tweet_json
```

Here we're creating a list of terms that we want to use for searching.

```python
search_terms = ['chicago','bar','lives']
```

This converts a list into a pipe-delimited (|) string. Regular Expressions (regex) sees a pipe as alternation, or OR.

```python
match_regex = "|".join(search_terms)
```

Here's our list of terms as a pipe-delimited string

```python
match_regex
```

Pandas uses 'masking' functionality to query dataframes. To use a mask, we need to have a series of boolean true|false values. If the corresponding entry is true, pandas will keep that entry in the returned list. If the corresponding entry is false, pandas will exclude it.

A mask looks just like a list of True or False values. We're seeing if any of the 'body' cells contain one or more of our list above. If the 'body' does, then it's value is True. If not, it's value is false.

We use the `.str` accessor to treat the values in the 'body' cells as strings. From there, we use the `.contains()` method to evaluate the body for whether or not it _contains_ the values in the first argument, **match_regex**. Since this is a 'regular expression' the | tells us check for one-or-more. The `flags=re.IGNORECASE` part tells regular expressions to ignore upper or lower case. Usually regular expressions are case sensitive, so we're turning that off. 

```python
tweet_json.body.str.contains(match_regex,flags=re.IGNORECASE)
```

We're going to take the previous line and put it in square brackets to the right of the dataframe variable name. This way the true/false values basically turn the corresponding lines in the dataframe on or off:

```python
tweet_json[tweet_json.body.str.contains(match_regex,flags=re.IGNORECASE)]
```
